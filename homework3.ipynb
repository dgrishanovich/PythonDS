```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.ensemble import RandomForestClassifier
```

```python
grid_search_cv_model = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                                    param_grid=parameters,
                                    scoring='roc_auc',
                                    cv=3)
```

```python
grid_search_cv_model.fit(X_train, y_train)
```

```python
grid_search_cv_model.best_params_
```

```python
y_pred_proba = grid_search_cv_model.predict_proba(X_test)[:, 1]
```

```python
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
auc_test = roc_auc_score(y_test, y_pred_proba)
auc_test
```
